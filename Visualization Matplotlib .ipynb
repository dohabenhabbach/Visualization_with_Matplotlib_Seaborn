## Line Chart

A line chart, also known as a line graph or line plot, is a type of data visualization used to display data points or values over a continuous interval or time period. Line charts are particularly useful for showing trends and changes in data over time.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
points=np.array([4,7,2,19,20])
```

```python
plt.plot(points)
```

```python
a2=np.array([1,2,3,4,5,6])
plt.plot(a2,linestyle='dotted')
plt.show()
```

## Bar Chart
A bar chart is a type of data visualization that represents data using rectangular bars. Bar charts are used to compare and display the values of different categories or groups. 

```python
x=np.array(['x','y','z','w'])
y=np.array([25,28,2,55])
```

```python
plt.bar(x,y)
```

```python
x1=np.array([20,7,66,22,31])
y1=np.array([2,9,10,22,67])
```

```python
plt.scatter(x1,y1)
plt.show()
```

## Pie Chart
A pie chart is a circular data visualization that is used to represent the distribution of a categorical variable as a "pie." Each category is represented as a slice of the pie, with the size of each slice proportional to the value it represents in the dataset. Pie charts are useful for displaying the relative proportions of different categories within a whole. 

```python
p =np.array([20,13,50,10])
plt.pie(p)
```

## Histogram
A histogram is a graphical representation of the distribution of a dataset. It displays the frequencies (or counts) of values within different bins or intervals. Histograms are useful for visualizing the underlying data distribution and identifying patterns

```python
h1 =np.random.normal(170,10,260)
plt.hist(h1)
```

## Box Plot
A box plot, also known as a box-and-whisker plot, is a data visualization that displays the distribution and spread of a dataset, as well as identifies potential outliers.

```python
df= sns.load_dataset('tips')
df
```

```python
df.boxplot(by='day',column=['total_bill'],grid=False)
```

## Dist Plot
A distribution plot, often created using the distplot function in Seaborn, is a data visualization that displays the distribution of a univariate dataset, showing the frequency of data points along a continuous axis. 

```python
titanic =sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
age_ =titanic['age'].dropna()
```

```python
sns.displot(age_,bins=30,kde=False)
```

## RegPlot
A regplot, short for "regression plot," is a data visualization that combines a scatter plot of data points with a linear regression line that best fits the data. It is used to explore the relationship between two variables and visualize how they correlate. 

```python
data=sns.load_dataset('mpg')
```

```python
data.head()
```

```python
sns.regplot(x='mpg',y='acceleration',data=data)
```
